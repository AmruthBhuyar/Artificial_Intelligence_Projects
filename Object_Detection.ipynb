````python
import cv2
import time
import imutils

cam = cv2.VideoCapture(0) # take 0 if primary camera/ 1 for secondary camera (usb camera)
time.sleep(1)

firstFrame = None
area = 500 

while True:
    _,img = cam.read() # _, variable is unknown take this.
    text = "Normal"
    img = imutils.resize(img, width = 500)
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gaussianImg = cv2.GaussianBlur(grayImg, (21,21), 0)
    if firstFrame is None:
        firstFrame = gaussianImg
        continue
    imgDiff = cv2.absdiff(firstFrame, gaussianImg)
    threshImg = cv2.threshold(imgDiff, 25, 255, cv2.THRESH_BINARY)[1]
    threshImg = cv2.dilate(threshImg, None, iterations = 2)
    cnts = cv2.findContours(threshImg.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x,y,w,h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0),2)
        text = "Moving object Detected."
        print(text)
    cv2.putText(img, text, (10,20),
        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2)
    
    cv2.imshow("cameraFeed1.jpg",img)
    key = cv2.waitKey(1)&0xFF  # 0 resembles continuous and 1 because returns 32 bit value and oxFF with and operation gives 8 bit data.
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()
````

````python
'''Explaination of program:
    

# Imports

import cv2
import time
import imutils
```
- **cv2**: OpenCV library for computer vision tasks.
- **time**: Used for adding delays.
- **imutils**: A series of convenience functions to make basic image processing operations easier.

# Initializing the Camera

cam = cv2.VideoCapture(0) 
time.sleep(1)
```
- **cv2.VideoCapture(0)**: Starts capturing video from the primary camera (0). Use 1 for a secondary camera.
- **time.sleep(1)**: Adds a delay of 1 second to allow the camera to warm up.

# Setting Initial Variables

firstFrame = None
area = 500

- **firstFrame**: Stores the first frame of the video.
- **area**: Minimum area of the detected contour to consider as motion.

# Main Loop

while True:
    _, img = cam.read()
    text = "Normal"
    img = imutils.resize(img, width=500)
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gaussianImg = cv2.GaussianBlur(grayImg, (21, 21), 0)

- **cam.read()**: Captures a frame from the camera.
- **imutils.resize(img, width=500)**: Resizes the frame to a width of 500 pixels.
- **cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)**: Converts the frame to grayscale.
- **cv2.GaussianBlur(grayImg, (21, 21), 0)**: Applies Gaussian blur to reduce noise.

# Initial Frame Handling

    if firstFrame is None:
        firstFrame = gaussianImg
        continue

- If the **firstFrame** is not set, it assigns the current blurred grayscale frame to **firstFrame** and continues to the next iteration.

# Frame Difference Calculation

    imgDiff = cv2.absdiff(firstFrame, gaussianImg)
    threshImg = cv2.threshold(imgDiff, 25, 255, cv2.THRESH_BINARY)[1]
    threshImg = cv2.dilate(threshImg, None, iterations=2)

- **cv2.absdiff(firstFrame, gaussianImg)**: Computes the absolute difference between the first frame and the current frame.
- **cv2.threshold(imgDiff, 25, 255, cv2.THRESH_BINARY)**: Converts the difference image to a binary image where the pixel values above 25 are set to 255 (white), and below are set to 0 (black).
- **cv2.dilate(threshImg, None, iterations=2)**: Dilates the binary image to fill in holes.

# Contour Detection

    cnts = cv2.findContours(threshImg.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)

- **cv2.findContours(threshImg.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)**: Finds the contours in the thresholded image.
- **imutils.grab_contours(cnts)**: Extracts the contours from the tuple returned by `cv2.findContours`.

# Contour Processing

    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x, y, w, h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        text = "Moving object Detected."
        print(text)

- Loops through each contour and checks if its area is larger than the specified threshold (500).
- **cv2.boundingRect(c)**: Computes the bounding box for the contour.
- **cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)**: Draws a rectangle around the detected motion in green.
- Updates the text to indicate motion is detected.

# Displaying Text and Image

    cv2.putText(img, text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    cv2.imshow("cameraFeed1.jpg", img)

- **cv2.putText(img, text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)**: Adds the status text to the frame.
- **cv2.imshow("cameraFeed1.jpg", img)**: Displays the frame with annotations.

# Exiting the Loop

    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()

- **cv2.waitKey(1) & 0xFF**: Waits for 1 millisecond for a key press. The `& 0xFF` operation ensures compatibility with 32-bit systems.
- If the 'q' key is pressed, the loop breaks.
- **cam.release()**: Releases the camera resource.
- **cv2.destroyAllWindows()**: Closes all OpenCV windows.

### Visual Representation
Here's a visual representation of the process:

Visual Representation
Here's a visual representation of the process:

1. Capture Frame:

2. Resize Frame:

3. Convert to Grayscale:

4. Apply Gaussian Blur:

5. Set Initial Frame (first iteration only):

6. Calculate Frame Difference:

7. Thresholding and Dilation:

8. Find and Process Contours:

9. Draw Bounding Box and Display:

10. Wait for Key Press and Exit:

This program effectively detects motion by analyzing the difference between the initial frame and subsequent frames, highlighting areas of movement with bounding boxes.'''
````
